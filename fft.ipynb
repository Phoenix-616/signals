```python
import numpy
import math
import matplotlib.pyplot as plt
```

```python
def genSin(steps, discretization):
    sinwave = []
    for i in range(0, steps):
        sinwave[len(sinwave):] = [math.sin(i * math.pi / discretization)]
    return sinwave
```

```python
sinwave = genSin(2000, 100)
sinwave[len(sinwave):] = genSin(2000,50)
plt.plot(sinwave)
plt.show
```

```python
a = numpy.fft.fft(sinwave)
print(a[0])
print(type(a[0]))
plt.plot(a)
plt.show
```

```python
b = numpy.fft.ifft(a)
plt.plot(b)
plt.show
```

```python
amp = numpy.abs(a)
plt.plot(amp)
plt.show
```

```python
ph = numpy.angle(a, deg=False)
plt.plot(ph)
plt.show
```
