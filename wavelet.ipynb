```python
import math
import matplotlib.pyplot as plt
import numpy as np
import pywt
```

```python
def genSin(steps, discretization):
    sinwave = []
    for i in range(0, steps):
        sinwave[len(sinwave):] = [math.sin(i * math.pi / discretization)]
    return sinwave
```

```python
a = genSin(500, 100)
a.extend(genSin(500, 10))
a.extend(genSin(500, 100))
plt.plot(a)
plt.show
```

```python
coef, freqs = pywt.cwt(a, np.arange(1, 100), 'morl')
```

```python
from mpl_toolkits import mplot3d

ax = plt.axes(projection='3d')
# ax.set_zscale('log')
# Data for three-dimensional scattered points
zdata = []
xdata = []
ydata = []
for i in range(len(freqs)):
    for j in range(len(coef[i])):
        xdata.append(i)
        ydata.append(j)
        zdata.append(coef[i][j])
plt.xlabel('Scale')
plt.ylabel('Steps')
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
```
